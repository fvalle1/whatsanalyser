```julia
using DataFrames
using PyPlot
```

```julia
include("whatsapp.jl")
using Main.whatsapp
```

```julia
fig, axs = PyPlot.subplots(1,2, figsize=(18,7))

for chat in ["_chat.txt" "_chat 2.txt" "_chat 3.txt" "_chat 4.txt" "_chat 5.txt"]
    voc, h, M , f = process_chat(chat)
    axs[1].plot(f, lw=10, label = String(chat))
    axs[1].set_xlim(1,length(f))
    axs[2].scatter(M,h, alpha=0.5, lw=5, s = 100)
end

ax=axs[1]
ax.plot([1. /x for x in 1:1000], ls="--", lw=15, c="gray")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("rank", fontsize=30)
ax.set_ylabel("frequency", fontsize=30)
ax.legend(fontsize=30)
ax.tick_params(labelsize=25, length=10)
ax.tick_params(which="minor", length=8)


ax = axs[2]

#ax.set_xscale("log")
#ax.set_yscale("log")
ax.set_xlabel("M", fontsize=30)
ax.set_ylabel("h", fontsize=30)
ax.tick_params(labelsize=25, length=10)
ax.tick_params(which="minor", length=8)

PyPlot.tight_layout()
PyPlot.show()
fig.savefig("Zipf_wa.pdf")
```
